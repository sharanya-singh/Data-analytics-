````python

import matplotlib.pyplot as plt
import numpy as np
from scipy import stats


#dataset: monthky temperatures (in C)
temperatures =np.array ([22, 25, 28, 30, 27, 26, 24, 23, 25, 22, 21, 20])

# Mean
mean = np.mean(temperatures)
print("Mean:", mean)

# Median
median = np.median(temperatures)
print("Median:", median)

# Mode
mode = stats.mode(temperatures)
print("Mode:", mode)

# Variance
variance = np.var(temperatures)
print("Variance:", variance)

plt.hist(temperatures, color = "yellow", edgecolor = "pink")
plt.hist(mode.mode, color='hotpink')

plt.xlabel('Temperature (C)')
plt.ylabel('Frequency')
plt.title('Temperature Data')
plt.show()
````

````python
#Bivariate Data Analysis

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression


#Data set of hours studied and exam scores
hours_studied= np.array([1, 2, 3, 4, 5, 6, 7]).reshape(-1,1)
exam_scores = np.array([45, 50, 55, 60, 65, 70, 75])

#Linear regression model
model = LinearRegression()
model.fit(hours_studied, exam_scores)

# Regression Coefficients
m = model.coef_[0]
c = model.intercept_
print(f"Regression equation: y = {m:.2f}x + {c:.2f}")

````

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
import seaborn as sns

df = pd.read_csv('/content/IRIS_data.csv')
df.head()
````

````python
df_setosa = df[df['species'] == 'setosa']
df_versicolor = df[df['species'] == 'versicolor']
df_virginica = df[df['species'] == 'virginica']
````

Linear plot

````python
df_setosa = df[df['species'] == 'Iris-setosa']
df_versicolor = df[df['species'] == 'Iris-versicolor']
df_virginica = df[df['species'] == 'Iris-virginica']
plt.plot(df_setosa['sepal_length'], np.zeros_like(df_setosa['sepal_length']),'o')
plt.plot(df_versicolor['sepal_length'], np.zeros_like(df_versicolor['sepal_length']),'o')
plt.plot(df_virginica['sepal_length'], np.zeros_like(df_virginica['sepal_length']),'o')
plt.legend(['setosa', 'versicolor', 'virginica'])
plt.xlabel('sepal_length')
````
